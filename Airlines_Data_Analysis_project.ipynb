Q.1. What are the airlines in the dataset, accompanied by their frequencies?**bold text**

```python
data['airline'].value_counts(ascending = True).plot.barh(color = ['Green', 'Blue'])
plt.title('Airline with Frequencies')
plt.xlabel('Frequency of flights')
plt.ylabel('Airline Names')
plt.show()

# Barh -- for making the bars horizontal
# Value_counts -- to find unique value with their counts
```

Q.2. Show Bar Graphs representing the Departure Time & Arrival Time

```python
plt.figure(figsize= (17,4))


# Chart number 1
plt.subplot(1,2,1)

plt.bar(data['departure_time'].value_counts().index, data['departure_time'].value_counts().values, color = ['Green', 'Yellow'])
plt.title('Departure Time')
plt.xlabel('Departure Timings')
plt.ylabel('Number of flights')

# Chart number 2
plt.subplot(1,2,2)
plt.bar(data['arrival_time'].value_counts().index, data['arrival_time'].value_counts().values, color = ['Red', 'Blue'])
plt.title('Arrival Time')
plt.xlabel('Arrival Timings')
plt.ylabel('Number of flights')


plt.show()

```

Q.3.  Show Bar Graphs representing the Source City & Destination City.

```python
plt.figure(figsize= (17,4))


# Chart number 1
plt.subplot(1,2,1)

plt.barh(data['source_city'].value_counts().index, data['source_city'].value_counts().values, color = ['green', 'pink'])
plt.title('Source city')
plt.xlabel('No. of flights')
plt.ylabel('City')

# Chart number 2
plt.subplot(1,2,2)
plt.barh(data['destination_city'].value_counts().index, data['destination_city'].value_counts().values, color = ['magenta', 'yellow'])
plt.title(' Destination City')
plt.xlabel('No. of flights')
plt.ylabel('City')


plt.show()
```

Q.4. Does price varies with airlines ?

```python
data.head()
```

```python
# grouping airlines and checking their mean prices
data.groupby('airline')['price'].mean()
```

```python
# plotting a categorical chart using seaborn library
sns.catplot(x='airline',y='price',kind='bar',hue='class',palette='rocket',legend=True,data=data)
plt.title('Variation in prices of Airlines')
plt.xlabel('Airline Names')
plt.ylabel('Prices')
plt.show()
```

Q.5. Does ticket price change based on the departure time and arrival time?

```python
data.head()
```

```python
# Checking mean ticket price based on departure time

data.groupby('departure_time')['price'].mean()
```

```python
# Checking mean ticket price based on arrival time

data.groupby('arrival_time')['price'].mean()
```

```python
sns.catplot(x = 'departure_time', y = 'price', kind = 'bar', data = data)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.catplot(x = 'arrival_time', y = 'price', kind = 'bar', data = data)
plt.xticks(rotation=90)
plt.show()
```

```python
# Relational plot for departuretime and arrival time
sns.relplot(x = 'arrival_time', y = 'price', data = data, col = 'departure_time', kind = 'line')
plt.show()
```

Q.9. What will be the Average Price of Vistara airline for a flight from Delhi to Hyderabad in Business Class ?

```python
data.head()
```

```python
# Filtering data based on the question asked

condition = data [ (data['airline'] == 'Vistara') & (data['source_city'] == 'Delhi') & (data['destination_city'] == 'Hyderabad') & (data['class'] == 'Business') ]
```

```python
# Mean / Average price based on using above filtering

result = condition['price'].mean()
print(result)
```

The Average Price of Vistara airline for a flight from Delhi to Hyderabad in Business Class : 47939.84 /-
